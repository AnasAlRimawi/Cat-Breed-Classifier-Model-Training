```python
import os
import csv
```

```python
folder_path = 'test/'
csv_path = 'test.csv'

# Define a dictionary to map class names to breed IDs
class_to_breed ={'Abyssinian': 0,
 'Bengal': 1,
 'Birman': 2,
 'Bombay': 3,
 'British': 4,
 'Egyptian': 5,
 'Maine': 6,
 'Persian': 7,
 'Ragdoll': 8,
 'Russian': 9,
 'Siamese': 10,
 'Sphynx': 11}
# Open the CSV file in write mode
with open(csv_path, 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['image', 'breed ID', 'classname'])

    # Loop through all the image files in the folder
    for filename in os.listdir(folder_path):
        if filename.endswith('.jpg') or filename.endswith('.png'):
            # Extract the classname from the filename
            classname = filename.split()[0]
            # Map the classname to a breed ID using the dictionary
            breed_id = class_to_breed[classname]
            # Write the image name, breed ID, and classname to the CSV file
            writer.writerow([filename, breed_id, classname])
```
